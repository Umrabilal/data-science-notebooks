# While loop or infinite loop

```python
# while loop or infinite loop
x = 10
y = 1

while x>y:
  print(f" I run because, {x} > {y} return {bool(x>y)}")
  break
  y += 1
  print(y)
else:
  print(f" I run because, {x} > {y} returns (bool{x>y}")
```
